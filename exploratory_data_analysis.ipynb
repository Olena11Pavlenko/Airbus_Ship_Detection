```python
import pandas as pd

import os
from PIL import Image
import matplotlib.pyplot as plt


# Define paths
train_images_dir = './images/train_v2'
train_csv_file = './train_ship_segmentations_v2.csv'
```

```python
image_shapes = {}

# Iterate over all the files in the folder
for file_name in os.listdir(train_images_dir):
    # Construct the full path to each image
    image_path = os.path.join(train_images_dir, file_name)

    # Open the image and get its shape
    image = Image.open(image_path)
    image_shape = image.size[::-1]

    # Count the occurrence of each image shape
    if image_shape in image_shapes:
        image_shapes[image_shape] += 1
    else:
        image_shapes[image_shape] = 1

print (image_shapes)
```

```python
# Visualise 5 images

#Get a list of all files in the folder
files = os.listdir(train_images_dir)

# Select the first five image files
image_files = files[:5]

# Iterate over the selected image files
for file_name in image_files:
    # Construct the full path to each image
    image_path = os.path.join(train_images_dir, file_name)

    # Load the image
    image = plt.imread(image_path)

    # Display the image
    plt.imshow(image)
    plt.show()
```
